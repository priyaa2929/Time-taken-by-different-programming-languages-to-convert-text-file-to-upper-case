## Time taken by different programming languages to convert text file to upper case

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x  = [1, 2, 3, 4, 5]
y1 = [50, 60, 70, 80, 90]
y2 = [52, 63, 74, 85, 96]
y3 = [54, 66, 78, 90, 102]
y4 = [60, 75, 95, 120, 150]
y5 = [65, 85, 105, 135, 185]
pics = sns.lineplot(x = x, y = y1, label='C', linewidth=1)
sns.lineplot(x = x, y = y2, label='C++', linewidth=1)
sns.lineplot(x = x, y = y3, label='Java', linewidth=1)
sns.lineplot(x = x, y = y4, label='Perl', linewidth=1)
sns.lineplot(x = x, y = y5, label='Python', linewidth=1)
pics.set_xlabel('File size(GB)')
pics.set_xlabel('Time Taken(sec)')
pics.set_title('Performance of different programming languages')
```

```python
sns.barplot(x = x, y = y1, label="C" , color = "g")
sns.barplot(x = x, y = y2, label='C++',color = "b")
sns.barplot(x = x, y = y3, label='Java',color = "c" )
sns.barplot(x = x, y = y4, label='Perl',color = "r")
pics = sns.barplot(x = x, y = y5, label='Python',color="m")
pics.set_xlabel('File size(GB)')
pics.set_ylabel('Time Taken(sec)')
pics.set_title('Performance of different programming languages')
plt.legend()
```

```python
sns.scatterplot(x = x, y = y1, label="C" , color = "g")
sns.scatterplot(x = x, y = y2, label='C++',color = "b")
sns.scatterplot(x = x, y = y3, label='Java',color = "c" )
sns.scatterplot(x = x, y = y4, label='Perl',color = "r")
pics = sns.scatterplot(x = x, y = y5, label='Python',color="m")
pics.set_xlabel('File size(GB)')
pics.set_ylabel('Time Taken(sec)')
pics.set_title('Performance of different programming languages')


plt.grid()
```

```python
size=['1GB','2GB','3GB','4GB','5GB']
time=[50, 60, 70, 80, 90]


plt.pie(time,labels=size, autopct='%.3f%%')

plt.title('Pie chart for C language pertaining to different file size')

plt.show()
```
